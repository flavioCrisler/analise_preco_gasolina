# Gerando a tabela de variação percentual

```python
# retornou o última preço reportado de cada ano
df_rio_gasolina.groupby(df_rio_gasolina.index.year).last()
```

```python
df_preco_medio_gasolina = df_rio_gasolina.groupby(df_rio_gasolina.index.year).agg({
    'preco_medio_revenda': np.mean
})
```

```python
# gerando a tabela de variação
tabela_variacao = round((df_preco_medio_gasolina.pct_change() * 100), 2)
tabela_variacao.columns = ['varicao_percentual']
```

```python
# Criando um novo DataFrame que será o valor anterio
preco_deslocado = df_preco_medio_gasolina.shift(1)
preco_deslocado.columns = ['valor_ano_anterior']
```

```python
# criando o DataFrame final que será a junção dos 3 DataFrames criando anteriormente
df_final = pd.concat([df_preco_medio_gasolina, preco_deslocado, tabela_variacao], axis=1)
```

```python
# para o resultado final é necessário dos resultados apenas de 2006 acima
tabela_final = round(df_final.loc[2006:], 2)
```
